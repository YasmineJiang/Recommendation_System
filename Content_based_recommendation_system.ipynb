```python
import pandas as pd
import numpy as np
```

```python
credits = pd.read_csv('tmdb_5000_credits.csv')
movies = pd.read_csv('tmdb_5000_movies.csv')
```

```python
credits.head()
```

```python
movies.head()
```

```python
credits_rename = credits.rename(index=str,columns={'movie_id':'id'})
merge_df = movies.merge(credits_rename,on='id')
merge_df.head()
```

```python
cleaned_df = merge_df.drop(columns=['homepage','title_x','title_y','status','production_countries'])
cleaned_df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfv = TfidfVectorizer(min_df=3,max_features=None,
                      strip_accents='unicode',analyzer='word',token_pattern='\w{1,}',
                      ngram_range=(1,3),
                      stop_words='english')

cleaned_df['overview'] = cleaned_df['overview'].fillna('')
```

```python
tfv_matrix = tfv.fit_transform(cleaned_df['overview'])
tfv_matrix
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel
sig = sigmoid_kernel(tfv_matrix,tfv_matrix)
```

```python
indices = pd.Series(cleaned_df.index,index=cleaned_df['original_title']).drop_duplicates()
indices
```

```python
def giv_rec(title,sig=sig):
  idx =indices[title]
  sig_scores = list(enumerate(sig[idx]))
  sig_scores = sorted(sig_scores,key=lambda x:x[1], reverse=True)
  sig_scores = sig_scores[1:11]
  movie_indices = [i[0] for i in sig_scores]

  return cleaned_df['original_title'].iloc[movie_indices]
```

```python
giv_rec('Avatar')
```
